```python
import pandas as pd
import sys, os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.rcdefaults()
```

```python
## test case 1
plt.ylim([0.0, 1.05])
class_idx = ['bass clef', 'half note', 'quarter note', 'quarter rest', 'g clef'];
numeric_acc = [0, 1, 1, 0, 1];

plt.bar(class_idx, numeric_acc, align='center', alpha=0.5, color = 'blue');
plt.xlabel(" ");
plt.ylabel('precision');
title_name = "test case 1";
plt.title(title_name);
plt.show()
'''
store_img = "../Output/Predict_Summary/0317_test/barplot_summary.png";
plt.savefig(store_img);
'''
## clear the canvas
plt.clf()
plt.cla()
plt.close()
```
