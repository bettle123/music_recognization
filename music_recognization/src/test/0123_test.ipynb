```python
# import necessary packages used in this project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
timestep = [1,2,3,4,5];
val_loss = [1000, 100, 20, 5, 1];
plt.plot(timestep,val_loss);
plt.xlabel('timestep')
plt.title("test")
plt.savefig("test_plot.png")
#plt.show()
```

```python
num_epoch = [1, 2, 3, 4, 5];
train_loss = [300, 200, 100, 50, 20];
test_acc = [0.15, 0.3, 0.4, 0.5, 0.6];

model_name = "../Model/0123_1/model_6_classes_0123_1";
basefilename = os.path.basename(model_name);

plt.figure(1)

plt.figure(1).suptitle(basefilename, fontsize=12)

plt.subplot(211)
plt.plot(num_epoch, train_loss);
plt.xlabel('# of epoch');
plt.title("Training Loss");

plt.subplot(212)
plt.ylim([0.0, 1.0])
plt.plot(num_epoch, test_acc);
plt.xlabel('# of epoch');
plt.title("Test Accuracy");
    
plt.subplots_adjust(top=0.80, bottom=0.1, left=0.10, right=0.95, hspace=1.0, wspace=0.35)
                    
#store = model_name+".png";
#plt.savefig(store);
```
