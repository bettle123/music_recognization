```python
import tensorflow as tf
import numpy as np
import os,glob
import sys,argparse
import cv2 as cv
```

```python
model_number = "0125_3";
model_dir = "../../Model/"+model_number+"/";
model_name = "model_5_classes_"+model_number+".meta";
## CNN restore the saved model 
sess = tf.Session();
## Recreate the model network graph.
model_total_dir = model_dir+model_name;
print("Model: ",model_total_dir);
saver = tf.train.import_meta_graph(model_total_dir)
## Load the weights saved using the restore method.
saver.restore(sess, tf.train.latest_checkpoint(model_dir))
## Access the default graph which we have restored
graph = tf.get_default_graph() 
```

```python
filename = "../../sample/white.png";
src = cv.imread(filename);
```

```python
crop_img = src;
## put crop images in the CNN model
image_size = 128;
num_channels=1
images = []
## resize the image to fit our CNN model
image = cv.resize(crop_img, (image_size, image_size),0,0, cv.INTER_LINEAR)
    
## Convert RGB into gray images
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
height, width = image.shape[:2]
image = np.reshape(image, (height, width, num_channels))

images.append(image)
images = np.array(images, dtype=np.uint8)
images = images.astype('float32')
images = np.multiply(images, 1.0/255.0) 
x_batch = images.reshape(1, image_size,image_size,num_channels)
```

```python
y_pred = graph.get_tensor_by_name("y_pred:0");
#print(y_pred.name)
#print(y_pred)
#@y_pred:0, 0 has meaning
# Tensor names must be of the form "<op_name>:<output_index>".
# Tensorflow adds the :0 to the name I specify
#print(y_pred)
x = graph.get_tensor_by_name("x:0");
#print(x.name)
#print(x)
y_true = graph.get_tensor_by_name("y_true:0");
#print(y_true.name)
#print(y_true)
```

```python
#print(x_batch)

y_test_images = np.zeros((1, 5));

feed_dict_testing = {x: x_batch, y_true: y_test_images};
batch_result = sess.run(y_pred, feed_dict=feed_dict_testing);
print(batch_result)

logits = graph.get_tensor_by_name("add_3:0");

logits_result = sess.run(logits, feed_dict=feed_dict_testing);



#batch_result_cls = sess.run(tf.argmax(y_pred, axis=1), feed_dict_testing)
#print(batch_result_cls)

#op = graph.get_operation_by_name('logits')


'''
batch_result_cls = sess.run(tf.argmax(y_pred, axis=1), feed_dict_testing)
#print(batch_result)
#print(batch_result_cls)
'''
'''
batch_result = sess.run(y_pred, feed_dict=feed_dict_testing);
        
## @result: the predict probability of each classes
result = batch_result[0];
print("result: ", result);
'''
```
